```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Abid/Desktop/Python/Abid/airbnb-data-analysis/data/listings.csv")

# Display basic info
df.head()
df.info()
df.describe()

# Check for missing values
df.isnull().sum()
# Drop duplicates
df = df.drop_duplicates()
df.rename(columns={"neighbourhood_group": "region", "price": "listing_price"}, inplace=True)
df["room_type"].value_counts().plot(kind="bar")

df["availability_365"].hist(bins=30)
plt.title("Availability of Airbnb Listings")
plt.xlabel("Days Available")
plt.ylabel("Count")
plt.show()



```
